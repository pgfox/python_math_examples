### summation
$$ \Sigma $$
$$ \Pi $$


```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
import math as math
import matplotlib.pyplot as plt

lst = [1,2,3,4,5]
print(f"the original list is {lst}")
print(f"sum of lst {sum(lst)}")
print(f"np.sum of lst {np.sum(lst)}")
print(f"np.cumsum is the product of the list {np.cumsum(lst)}")
print(f"np.prod is the product of the list {np.prod(lst)}")

```

```python
plt.plot(lst, label="orig")
plt.plot(np.cumsum(lst), label="cumsum")
plt.legend()
plt.show()
```

$$\frac{\Sigma a}{\Sigma a^2} \quad = \quad \frac{1}{\Sigma a} \quad ???$$

```python
#create an np array
the_list = np.arange(1,6)

print(f"the_list^2={the_list**2} ")
lhs = sum(the_list) / sum(the_list**2)
rhs = 1/sum(the_list)

print(f"lhs={lhs} and rhs={rhs}")
print(f"lhs = rhs ? {lhs==rhs}")


```

$$\frac{\Pi a}{\Pi a^2} \quad = \quad \frac{1}{\Pi a} \quad ???$$

```python
lhs = np.prod(the_list)/np.prod(the_list**2)
rhs = 1/np.prod(the_list)
print(f"lhs={lhs} and rhs={rhs}")
print(f"lhs = rhs ? {lhs==rhs}")
```

### differences

$$ y_t = x_{t+1} - x_t $$

You are subtracting the following value (t+1) from the currect value (t)


```python
x = [1,10,30,100,-70]
my_diff = np.diff(x)
print(my_diff)
```

```python
the_list = np.arange(100,111)
print(np.diff(the_list))
# second order discrete difference
# second order - it takes the difference of the difference
print(np.diff(the_list,2))
#same as doing
print(np.diff(np.diff(the_list)))
```

```python
x = np.linspace(-2,2,101)
f_x = x**2
g_x = x**6

df = np.diff(f_x)
dg = np.diff(g_x)
dx = np.diff(x)

    
plt.plot(x,f_x,'b', label='$ f(x)=x^2 $')
plt.plot(x,g_x, 'r', label='$ g(x)= X^6 $')

# rate of change.
plt.plot(x[:-1],df/dx,'b--', label="df/dx")
plt.plot(x[:-1],dg/dx,'r--', label="dg/dx")

plt.axis([-2,2,-1,1])
plt.grid()
plt.legend()

plt.axis([-4,4,-4,4])

#shade x,y axis lines
#get current axis (for this plot)
the_axis = plt.gca()
plt.plot(the_axis.get_xlim(),[0,0],'k')
plt.plot([0,0],the_axis.get_ylim(),'k')
plt.legend()

plt.show()
```

### polynomials

$$ a_0 + a_1x + a_2x^2 + .... + a_nx^n $$

a is coeficients 

roots - equation is equal to zero  and solve for x


```python
#it needs to be highest degree first 
#x^2 + 5x + 6
my_coeffs = [1,5,6]
the_answer = list(np.roots(my_coeffs))
display(the_answer)
print(f"x= {the_answer[0]} and x={the_answer[1]}")
```

```python
#it needs to be highest degree first 
#x^2 - 9
my_coeffs = [1,0,-9]
the_answer = list(np.roots(my_coeffs))
display(the_answer)
print(f"x= {the_answer[0]} and x={the_answer[1]}")
```

```python
#it needs to be highest degree first 
#4x^3 + 6x^2 - 6x -9= 0
my_coeffs = [4,6,-6,-9]
the_answer = list(np.roots(my_coeffs))
display(the_answer)
print(f"x= {the_answer[0]} and x={the_answer[1]} and x={the_answer[2]}")
```

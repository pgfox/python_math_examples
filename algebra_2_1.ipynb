### unit circle

```python
rads = np.linspace(0,2*np.pi,360)

#plotting the x/radius ratio, y/radius ratio up until 2.PI 
# will create the unit circle. 
plt.plot(np.cos(rads),np.sin(rads),'red')

# make a larger circle from unit circle - use a scalar
plt.plot(3*np.cos(rads),3*np.sin(rads),'blue')


plt.axis([-5,5,-5,5])

#shade x,y axis lines
#get current axis (for this plot)
the_axis = plt.gca()
plt.plot(the_axis.get_xlim(),[0,0],'k')
plt.plot([0,0],the_axis.get_ylim(),'k')
plt.grid()
plt.axis('square')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()

```

### plot a circle

```python
x = np.linspace(-5,5,200)

r = x[-1]
y_upper = np.sqrt(r**2 - x**2)
y_lower = - y_upper

plt.plot(x,y_upper,'green')
plt.plot(x,y_lower,'green')

the_axis = plt.gca()
plt.plot(the_axis.get_xlim(),[0,0],'k')
plt.plot([0,0],the_axis.get_ylim(),'k')
plt.grid()
plt.axis('square')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

### natural exponent and logarithm

$$ y = e^x $$

$$ y = log(x) $$

These are inverse functions of each other .
So the output of the first used as the input to the second will return the first input. 


```python
x = np.linspace(0.1,10,80)
y = np.exp(x)
display(Math("$$y = e^x"))
plt.plot(x,y,'green')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

```python
x = np.linspace(0.1,10,80)
#print(x)
y = np.log(x)
#print(y)
plt.plot(x,y,'green')
display(Math("y=log(x)"))
plt.grid()
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()

```

```python
x = np.linspace(0.1,2,80)
y1 = np.exp(x)
y2 = np.log(x)

plt.plot(x,y1,'green',label='$e^x$')
plt.plot(x,y2,'blue',label="log(x)")
plt.legend()
plt.grid()
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

### guassian 

$$ y = e ^ \frac{-4 ln (2) x ^2}{h^2}$$

x = is the x axis

h = full width at half maximum. h is fixed for the guassian and x changes. 

Alternatively written as :
$$ y = exp \left( \frac {-4ln(2)x^2}{h^2}\right) $$




```python
#plt.rc_context( {'xtick.color':'white','figure.facecolor':'white'})

h=1
#h=2
x = np.linspace(-5,5,100)
y = np.exp((-4*np.log(2)*x**2)/h**2)

plt.plot(x,y)
plt.plot([-5,5],[0.5,0.5],"--")
plt.plot([-0.5,-0.5],[0.5,0],"g--")
plt.plot([0.5,0.5],[0.5,0],"g--")
plt.grid()
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

### linear and log scales

linear    =  1,   2,  3,     4,  5
log-space =  1, 1.5,  2.2, 3.3,  5

linear -  the distance between each increment is the same . 

log space - the distance is not the same. Distance is increasing . The distance between 3.3 and 5 is 3 time larger than 1 and 1.5

log scale maintaines its proportionality (multiply by x) not the actual distance.

log scale to base 2 => 2 to 3 in linear terms is 4 to 8 ==> each step multiplies by 2
log scale to base 10 => 2 to 3 in linear termes is 100 to 1000 ==> each step multiplies by 10 




```python
# remember linspace includes the 'start' and it divides the end number by num -1
my_linear_space = np.linspace(0,10,11)
print(my_linear_space)
```

```python
# it creates an array with 2^0, 2^1, 2^2, 2^2
my_log_space = np.logspace(0,5,6,base=2)
print(my_log_space)

plt.yscale('log')
plt.plot(my_log_space,'b')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

### logarithm properties

$$ log(a \times b) = log(a) + log(b)  $$



```python
print(f"log(base 10) of 100*10 = {np.log10((100 * 10))}" )
print(f"log(base 10) of 100  = {np.log10(100)} and the log(base 10) of 10 is {np.log10(10)}" )
```

$$ log \left( \frac {a}{b} \right) = log(a) - log(b)$$

```python
print(f"log(base 10) of 1000/10 = {np.log10((1000 / 10))}" )
print(f"log(base 10) of 100  = {np.log10(1000)} and the log(base 10) of 10 is {np.log10(10)}" )
print(f"log(base 10) of 100  - log(base 10) of 10 => {np.log10(1000) - np.log10(10)}" )
```

$$ log(a^b) = b log(a)$$

```python
print(f"log10(1000) = {np.log10(1000)}")
print(f"log10(1000^2) = {np.log10(1000**2)}")
```

### geometric sequence

 Geometric Sequence - each term is found by multiplying the previous term by a constant.
 
 Arithmetic Sequence -  every term is obtained by adding a fixed number (positive or negative or zero) to its previous term. 

```python
a = 3
c = 5
arith = a + c*np.arange(1,5)  
print(arith)
```

```python
a = 3
c = 5
geom = a * c**np.arange(1,5)  
print(geom)
```

```python
plt.plot(arith,'b')
plt.plot(geom,'g')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.grid()
plt.show()
```

### minima/maxima

finding minimum/maximum of a list 

```python
my_list = [10,14,16,8,7]

print(f"max {np.max(my_list)}")
print(f"max position {np.argmax(my_list)}")
print(f"min {np.min(my_list)}")
print(f"min position {np.argmin(my_list)}")
```

```python
x = np.linspace(-5,5)
y= x**4 + x**2

y_min = np.min(y)
y_min_loc = np.argmin(y)
plt.plot(x,y,'ro-')
plt.axis([-1,1,-5,5])
plt.plot(x[y_min_loc],y_min,'bs')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

```python
import scipy.signal as sig

x = np.linspace(0,39,100)
y = -np.cos(x) - np.sqrt(x)
the_peaks = sig.find_peaks(y)
print(f"peaks found {the_peaks}")


plt.plot(x,y,'g-')
plt.plot(x[the_peaks[0]],y[the_peaks[0]],'bs')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

### even and odd functions

f(x) is odd if f(-x) = - f(x)

### random numbers 

using a **_normal distribution_** - drawing numbers at random you are more likely to get a zero and less likely the further away from zero. 


using a **_uniform distribution_** - equally likely to get numbers that fall between the value of 0 and 1 




```python
#get a random number
print(np.random.rand())

#get 5 random numbers
print(np.random.rand(5))
my_rands = np.random.rand(500)
plt.plot(my_rands,'g.')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()

#shows the distribution
plt.hist(my_rands)
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

```python
#random number in range 40-60
my_rands = np.random.rand(1000)
my_rands = my_rands * (20) + 40

plt.plot(my_rands,'b.')
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.axis([0,500,0,80])
plt.show()
```

```python
#normal distributed numbers
my_rands_norm = np.random.randn(5000)
plt.plot(my_rands_norm,'r.',alpha=0.5)
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()

#shows the distribution
plt.hist(my_rands_norm)
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.show()
```

 changing the standard deviation and mean of the normal distribution of the randomly generated numbers

```python
my_rands_norm = np.random.randn(500)

sigma = 5
mu = 10
my_rands_norm = sigma * my_rands_norm + mu
plt.hist(my_rands_norm)
plt.gca().tick_params(axis='x', colors='orange')
plt.gca().tick_params(axis='y', colors='orange')
plt.grid()
plt.show()
```

```python
# setting emperical mean to exactly 0
my_rands_norm = np.random.randn(500)
print(f"empirical mean BEFORE manipulation is {np.mean(my_rands_norm)}")
my_rands_norm = my_rands_norm - np.mean(my_rands_norm)
print(f"empirical mean AFTER manipulation is {np.mean(my_rands_norm)}")

# setting empirical std to exactly one
print(f"empirical standard deviation BEFORE manipulation is {np.std(my_rands_norm)}")
my_rands_norm = my_rands_norm / np.std(my_rands_norm)
print(f"empirical standard deviation AFTER manipulation is {np.std(my_rands_norm)}")
```

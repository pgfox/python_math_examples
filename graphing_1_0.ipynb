```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np
from IPython.display import display, Math
import math as math
```

### plotting points

```python
x=4
y=6

cordinates = [(1,1), (2,1), (3,2), (4,2), (5,3),(6,3)]

plt.plot(x,y,'rs')
plt.plot(1,3,'gs')

for i, cordin in enumerate(cordinates):
    plt.plot(cordin[0],cordin[1], 'o', label="index %s" %i)

plt.axis('square')
plt.axis([-7,7,-7,7])
plt.grid()
plt.legend()
plt.show()
```

```python
x = sym.symbols('x')

y = x**2 - 3*x

for i in range(-10,11):
    plt.plot(i, y.subs(x,i),'o')
    
plt.grid()
#set x and y axis
plt.axis([-15,15,-10,150])
plt.xlabel("label x")
plt.xlabel("label y")
plt.show()

```

### plotting a line

```python
cord_1 = (-5,-3)
cord_2 = (4,5)
cord_3 = (3,-2)

plt.axis([-6,6,-6,6])

#shade x,y axis lines
#get current axis (for this plot)
the_axis = plt.gca()
plt.plot(the_axis.get_xlim(),[0,0],'k')
plt.plot([0,0],the_axis.get_ylim(),'k')

#x cords, y cords
plt.plot([cord_1[0],cord_2[0]],[cord_1[1],cord_2[1]],'g')

# rgb color and thickness
plt.plot([cord_1[0],cord_3[0]],[cord_1[1],cord_3[1]],color=[.5,0.2,0.3],linewidth=3)
plt.grid()
plt.show()
```

```python


for x in range(0,75):
    x_cord = -(x-5)
    y_cord = (x-40)
    plt.plot([0,x_cord],[0,y_cord])

plt.grid()
plt.show()
```

### line - slope intercept form 

```python
#numpy way
x1 = np.array([-3,7])
m1 = 1
b1 = 0
y1 = m1 * x1 + b1

x2 = np.array([-3,7])
m2 = -5
b2 = -2
y2 = m2* x2 + b2

plt.axis([-10,10,-10,10])

#shade x,y axis lines
#get current axis (for this plot)
the_axis = plt.gca()
plt.plot(the_axis.get_xlim(),[0,0],'k')
plt.plot([0,0],the_axis.get_ylim(),'k')

plt.plot(x1,y1,'g', label='y1')
plt.plot(x2,y2,'r', label='y2')
plt.legend()
plt.grid()
plt.show()

```

### plot a polynomial

```python
x_values = range (-5,6)
y_values = np.zeros(len(x_values))

for i, x in enumerate(x_values):
    y_values[i] = x**2 - 2*x -3
    
plt.plot(x_values,y_values)
plt.grid()

plt.axis([-10,10,-10,10])

#shade x,y axis lines
#get current axis (for this plot)
the_axis = plt.gca()
plt.plot(the_axis.get_xlim(),[0,0],'k')
plt.plot([0,0],the_axis.get_ylim(),'k')

plt.show()
```

```python
#lbound and ubound - number of spaces
x_values= np.linspace(-4,4,50)
y_values = x_values **2 + 5*x_values 
y_values_2 = x_values **3 

plt.plot(x_values,y_values,label='$y_1 = x^2+5x$')
plt.plot(x_values,y_values_2, label='$y_2 = x^3$')
plt.grid()

plt.axis([-4,4,-10,10])

#shade x,y axis lines
#get current axis (for this plot)
the_axis = plt.gca()
plt.plot(the_axis.get_xlim(),[0,0],'k')
plt.plot([0,0],the_axis.get_ylim(),'k')
plt.legend()

plt.show()
```
